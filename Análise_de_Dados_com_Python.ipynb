```python
import pandas as pd

dados = {
    'Data_Vendas': ['2025-05-01', '2025-05-03', '2025-05-05',
                    '2025-05-07', '2025-05-08'],
    'Produto': ['Carrinho', 'Boneca', 'Jogo de Tabuleiro', 'Quebra-cabeça',
                'Bola'],
    'Categoria': ['Carros', 'Bonecas', 'Games', 'Jogos', 'Esportes'],  # corrigido
    'Quantidade': [2, 10, 20, 5, 6],
    'preco_unitario': [10.00, 20.00, 50.00, 15.00, 40.00]
}

df = pd.DataFrame(dados)
df.to_csv('vendas_brinquedos.csv', index=False)
df.head()
```

```python
df['Valor_Total'] = df['Quantidade'] * df['preco_unitario']
df.head()
```

```python
media_valor_total = df['Valor_Total'].mean()
print("A média é: ", media_valor_total)
```
